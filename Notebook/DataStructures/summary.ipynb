# Flat Sequences and container sequences

# Flat Sequences:
- Very fast
- commonly used for storing atomic data
    - such as numbers, bytes, chars

# Container Sequences
- More flexible
- Can show some unpredictable behaviours when used for storing mutable objects
    - i.e. nested data structures
- some objects contain references to other objects these
    are called containers


## List comprehensions and generators are powerful notation to build and initialize sequences

# Built-in sequences
## _Container sequences:_
- list, tuple and collections.deque
    - ALL can hold item of different types
    - Hold references to the object they contain

## _Flat sequences_
- str, bytes, bytearray, memory-view and array.array
    - Hold items of one type
    - physically stores the value of each item within its own memory space
      and not as distinct objects.
    - more compact but limited to holding primitive values
      like numbers, chars and bytes

 - list, bytearray, array.array, collections.deque and memory-view

## _Immutable Sequences:_
 - tuple, str, bytes


# List comprehension and generator expressions
one of the core reasons to adopt list comps over list is because of
it's clear readability nature take for instance.

    nums = '1234567'
    number_list = []
    for num in nums:
        number_list.append(int(num))
    print(number_list)
    [1, 2, 3, 4, 5, 6, 7]

Equivalently: This code below is much readable and faster

    nums = '1234567'
    number_list = [int(num) for num in nums]
    print(number_list)
    [1, 2, 3, 4, 5, 6, 7]

### _using star * to grab excess items_
Defining function parameters with *args to grab arbitrary excess arguments is a
classic python future
In python 3 it was extended to apply to parallel arguments as well

    a, b = *rest = range(5)
    a, b, rest
    (0, 1, [2, 3, 4])
